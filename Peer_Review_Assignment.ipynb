# Statistical tests for Hypothesis testing

```python
import scipy.stats
import statsmodels.api as sm
```

### Null Hypothesis : There is no significant difference in median value of houses bounded by the Charles river

```python
scipy.stats.levene(boston_df[boston_df['CHAS'] == 1]['MEDV'],
                   boston_df[boston_df['CHAS'] == 0]['MEDV'], center='mean')
```

since the p-value is less than 0.05 we can not assume equality of variance

```python
scipy.stats.ttest_ind(boston_df[boston_df['CHAS'] == 1]['MEDV'],
                      boston_df[boston_df['CHAS'] == 0]['MEDV'], equal_var = False)
```

**Conclusion:** Since the p-value is less than alpha value 0.05, we reject the null hypothesis as there is enough proof that there is a statistical difference  in median value of houses bounded by the Charles river.

### Null Hypothesis : There is no difference in Median values of houses (MEDV) for each proportion of owner occupied units built prior to 1940 (AGE)

```python
from statsmodels.formula.api import ols
lm = ols('MEDV ~ AGE', data = boston_df).fit()
table= sm.stats.anova_lm(lm)
print(table)
```

**Conclusion:** We can see from ANOVA since the p-value is less than 0.05, we will reject the null hypothesis. There is significant evidence that there a difference in Median values of houses (MEDV) for each proportion of owner occupied units built prior to 1940.

### Null Hypothesis : There is no relationship between Nitric oxide concentrations and proportion of non-retail business acres per town

```python
corelation_statistic,p_value = scipy.stats.pearsonr(boston_df['NOX'], boston_df['INDUS'])
print("corelation_Statistic: {0}, P-Value: {1}".format(corelation_statistic,p_value))
```

**Conclusion:** Since the p-value < 0.05, we reject  the Null hypothesis and conclude that there  exists a relationship between Nitric oxide concentrations and proportion of non-retail business acres per town.


### Null Hypothesis : There is no impact of an additional weighted distance  to the five Boston employment centres on the median value of owner occupied homes

```python
X = boston_df['DIS']
y = boston_df['MEDV']
X = sm.add_constant(X) 
model = sm.OLS(y, X).fit()
predictions = model.predict(X)
model.summary()
```

**Conclusion:** p < 0.05 there is evidence of correlation between additional weighted distance  to the five Boston employment centres on the median value of owner occupied homes.
